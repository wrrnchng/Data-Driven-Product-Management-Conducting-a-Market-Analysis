## Dataset: Individual Household Electric Power Consumption

**Donated on:** 8/29/2012

**Description:** Measurements of electric power consumption in one household with a one-minute sampling rate over a period of almost 4 years. Different electrical quantities and some sub-metering values are available.

### Data Used for Regression
- **Aggregated Data:** Daily grain.
- **Columns Used:**
  - `date`
  - `power_consumption` (output variable derived from original column: `Global_active_power`)
- **Preprocessing Steps:**
  - Removed rows with missing values.
  - Aggregated daily power consumption by summing `Global_active_power`.
  - Split data into training and testing sets based on the date.
  - Created new temporal features from date column: `year`, `semester`, `quarter`, `day_in_week`, `week_in_year`, `day_in_year`, and `month`.
